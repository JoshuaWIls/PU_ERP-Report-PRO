```python
#Task 1 code
import pandas as pd  
import numpy as np
from openpyxl import load_workbook  
data = pd.read_excel("SoCSE.xlsx")  
batch = data["Batch"].unique()
for i in batch:
    a = data[data["Batch"].str.contains(i)]  
    a.to_excel(i +"-xlsx.xlsx",index=False)
    wb = load_workbook(i+"-xlsx.xlsx")  
    ws = wb.active
    ws.insert_rows(1)   
    wb.save(i + "-xlsx.xlsx")  
    c = ws.cell(row=1, column=1)  
    c.value = i
    wb.save(i + "-xlsx.xlsx")
```

```python
#Task 2 and 3 code
import pandas as pd
from openpyxl import Workbook
hod_data = pd.read_excel("HoD-wise Sections.xlsx")
socse_data = pd.read_excel("SoCSE.xlsx")
workbook = Workbook()
workbook.remove(workbook.active)
hod_names = hod_data["HoD Name"].unique()
column_names = socse_data.columns.tolist()
for hod_name in hod_names:
    hod_sections = hod_data[hod_data["HoD Name"] == hod_name]["Sections"].tolist()
    for section in hod_sections:
        worksheet = workbook.create_sheet(title=section)
        worksheet.append([section]) 
        worksheet.append(column_names) 
        section_data = socse_data[socse_data["Batch"] == section]
        for index, row in section_data.iterrows():
            worksheet.append(list(row))
    workbook.save(f"{hod_name}.xlsx")
```
